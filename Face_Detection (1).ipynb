```
import cv2
from google.colab.patches import cv2_imshow
```

Providing haarcascade_frontalface_default.xml Dependencies file to CascadeClassifier.

Providing the image to Opencv imread function

Color Conversion using cvtColor

Using detectMultiScale for face detection.

```
 face_cascade= cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
img = cv2.imread('/content/1.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray,1.3,5)
print(faces)
```

Drawing the rectangle For showing the detected face.

```

for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
  rg=gray[y:y+h,x:x+w]
  rc=img[y:y+h,x:x+w]

cv2_imshow(img)


```
